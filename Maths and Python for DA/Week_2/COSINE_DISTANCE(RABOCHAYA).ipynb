# Сходство текстов

```python
import re
from collections import Counter
import numpy as np
from scipy.spatial.distance import cosine
```

```python
with open ('sentences.txt') as fin:
    data_list = fin.readlines()
```

```python
for i in range (len(data_list)):
    data_list[i] = data_list[i].lower()
```

```python
data_list
```

```python
print re.split('[^a-z]', data_list[0])
```

```python
list_of_words = []
for i in range (len(data_list)):
    list_of_words.extend(re.split('[^a-z]', data_list[i]))
```

```python
list_of_words
```

```python
for i in range (list_of_words.count('')):
    list_of_words.remove('')
```

```python
dict_of_words = Counter()
for word in list_of_words:
    dict_of_words[word] += 1
```

```python
new_dict_of_words = {}
for i in range (len(dict_of_words.keys())):
    new_dict_of_words[i] = dict_of_words.keys()[i]
dict_of_words = new_dict_of_words
```

```python
dict_of_words
```

```python
matrix = np.ones((len(data_list), len(dict_of_words.keys())))
```

```python
for i in range (len(data_list)):
    for j in range (len(dict_of_words.keys())):
        matrix[i, j] = re.split('[^a-z]', data_list[i]).count(dict_of_words[j])
```

```python
ans = []
for i in range (len(data_list)-1):
    answer = cosine(matrix[0,:], matrix[i + 1,:])
    ans.append(answer)
```

```python
ans
```

```python
print np.argsort(ans)[:2] + 1
```

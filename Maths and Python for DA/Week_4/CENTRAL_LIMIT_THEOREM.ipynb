```python
import numpy as np
import pandas as pd
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
%matplotlib inline
import matplotlib.pylab as plt
```

# Работа с распределением
### Выбираем непрерывное распределение

```python
norm_rv = sts.norm(0, 1)
```

### Генерируем выборку объемом 1000 из этого распределения

```python
sample = norm_rv.rvs(1000)
```

### Смотрим на теоретическую плотность вероятности и функцию распределения данного распределения

```python
x = np.linspace (-10, 10, 100000)
cdf = sts.norm.cdf(x)
pdf = sts.norm.pdf(x)
```

### Строим гистограмму выборки и функцию плотности вероятности распределения

```python
plt.hist (sample, label = 'sample', normed = True)
plt.plot (x, pdf, label = 'pdf')
plt.xlabel ('x')
plt.ylabel ('f(x)')
plt.legend ()
```

# Центральная предельная теорема
Дисперсия выборочного среднего равна 1/n, где n - количество элементов в выборке, среднеквадратичное отклонение - корень из дисперсии передаем в качестве аргумента в функцию плотности вероятности

```python
means = []
for _ in range (1000):
    means.append(np.mean(norm_rv.rvs(5)))
plt.hist (means, normed = True, label = 'sample means')
plt.plot (x, sts.norm.pdf(x, 0, np.sqrt(1.0/5)), label = 'pdf')
plt.xlabel ('x')
plt.ylabel ('f(x)')
plt.legend()
```

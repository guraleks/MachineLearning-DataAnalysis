```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as sts
%matplotlib inline
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
norm_rv = sts.norm(0, 1)
```

```python
x = np.linspace(-10, 10, 100000)
cdf = norm_rv.cdf(x)
plt.plot(x, cdf)
plt.xlabel('x')
plt.ylabel('F(x)')
```

```python
plt.plot(x, norm_rv.pdf(x))
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
sample = norm_rv.rvs(100)
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label = 'ECDF')
plt.plot(x, cdf, label = 'cdf')
plt.xlabel('x')
plt.ylabel('F(x)')
```

```python
type(sample)
```

```python
plt.hist(sample, normed = True, label = 'sample')
plt.xlabel('x')
plt.ylabel('number of samples')
plt.plot(x, norm_rv.pdf(x), label = 'theoretical pdf', alpha = 0.5)
plt.legend()
```
